```python
f=open("US_Declaration.pdf","rb")
```

```python
pdf_text=[]
```

```python
pdf_reader=PyPDF2.PdfReader(f)
```

```python
for p in range(len(pdf_reader.pages)):
    page=pdf_reader.pages[p]
    pdf_text.append(page.extract_text())
f.close()
```

```python
len(pdf_text)
```

```python
for f in pdf_text:
    print(f)
    print("\n")
    print("\n")
```
