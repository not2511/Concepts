```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
#create data frame
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df["target"]=iris.target
df.head()
```

```python
iris.target_names
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
#generating new column from existing one
df
```

```python
from matplotlib import pyplot as plt
```

```python
#separate 3 data species into different data frames
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='red')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal length (cm)'],color='blue')
plt.scatter(df1['petal length (cm)'],df1['petal length (cm)'],color='red')
```

```python
 from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flower_name'],axis='columns')
x.head
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```
